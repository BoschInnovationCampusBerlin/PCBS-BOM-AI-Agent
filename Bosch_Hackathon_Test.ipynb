# Table conversion

```python
import pandas as pd
import json

def summarize_component_offers(json_file_path: str, target_quantity: int):
    """
    Loads electronics component data from a JSON file and generates a
    summary table of available offers for each component.

    Args:
        json_file_path (str): The path to the JSON file containing component data.
    """
    try:
        with open(json_file_path, 'r', encoding='utf-8') as f:
            data = json.load(f)
    except FileNotFoundError:
        print(f"Error: JSON file not found at {json_file_path}")
        return
    except json.JSONDecodeError:
        print(f"Error: Could not decode JSON from {json_file_path}")
        return

    similar_parts = data.get('data', {}).get('supSearchMpn', {}).get('results', [])

    if not similar_parts:
        print("No similar parts data found in the JSON.")
        return

    for part_entry in similar_parts:
        part_details = part_entry.get('part', {})
        similar_components = part_details.get('similarParts', [])

        for component in similar_components:
            component_name = component.get('name', 'N/A')
            sellers_data = component.get('sellers', [])

            offers_list = []
            for seller in sellers_data:
                seller_name = seller.get('company', {}).get('name', 'N/A')
                seller_country = seller.get('country', 'N/A')
                offers = seller.get('offers', [])

                for offer in offers:
                    prices = offer.get('prices', [])
                    for price_info in prices:
                        min_order_quantity = price_info.get('quantity', 'N/A')
                        price = price_info.get('convertedPrice', 'N/A')

                        price_for_target_qty = price * max(min_order_quantity, target_quantity)

                        offers_list.append({
                            "Seller name": seller_name,
                            "Seller country": seller_country,
                            "Minimum order quantity": min_order_quantity,
                            "Price": price,
                            "Price for target quantity": price_for_target_qty
                        })

            if offers_list:
                df = pd.DataFrame(offers_list)
                df.sort_values(by=["Price for target quantity"], inplace=True)
                print(f"\n\n# Component: {component_name}")
                print("## Best offer:")
                print(df.iloc[0, :].to_string())
                print(f"\n## Total number of sellers: {len(sellers_data)}")
                #print(df.to_string(index=False))
            else:
                print(f"\n--- No offers found for Component: {component_name} ---")

# Specify the path to your JSON file
json_file_path = r"c:\Users\Daniel\Documents\Programming\VSCode Projects\agentic_ai_thesis\first_test_crew\test3_only_convprice.json"

# Call the function to process the data
summarize_component_offers(json_file_path, 1000)
```
